```python
import numpy as np
import matplotlib.pyplot as plt

from cobaya.run import run

from getdist.mcsamples import MCSamplesFromCobaya
from getdist import plots, MCSamples

from datetime import datetime
from skyfield.api import load, utc

plt.rcParams['text.usetex']=True
plt.style.use('classic')

plt.rcParams['font.size'] = 18
```

```python
# Create a timescale and ask the current time.
ts = load.timescale()


# Load the JPL ephemeris DE421 (covers 1900-2050).
#https://ui.adsabs.harvard.edu/abs/2019ascl.soft07024R
planets           = load('de421.bsp')
earth,sun,jupiter = planets['earth'],planets['sun'],planets['jupiter BARYCENTER']



#Data is collected from 10.1016/j.astropartphys.2022.102778
data              = np.loadtxt('./Eccentricity_Data_Fig4.txt')

#Gathering data from 03 october 2019 to 02 october 2021 :
data  = data[data[:,0]>2853]
```

```python
#preparing the distances

date = datetime(2011,12,11,0,0,0,tzinfo=utc)
t0   = ts.utc(date)
t    = np.arange(0,data[-1,0],0.5)
dtheory_sun = np.zeros(t.shape[0])
dtheory_jup = np.zeros(t.shape[0])

for i,dt in enumerate(t):
    tstep = t0 + dt
    
    astrometric_sun    = earth.at(tstep).observe(sun)
    lat, lon, distance = astrometric_sun.radec()
    dtheory_sun[i]     = distance.m
    
    astrometric_jupiter = earth.at(tstep).observe(jupiter)
    lat, lon, distance  = astrometric_jupiter.radec()
    dtheory_jup[i]      = distance.m
    
a_sun     = 1.496e11 #acoording to 1 AU
a_jupiter = 5*a_sun


dtheory_jup = dtheory_jup/a_jupiter
dtheory_sun = dtheory_sun/a_sun


rmean_sun = np.mean(1/dtheory_sun**2)
rmean_jup = np.mean(1/dtheory_jup**2)
```

```python
#prediction of the data is given below

rpredict_sun = np.zeros(data.shape[0])
rpredict_jup = np.zeros(data.shape[0])

for i,dt in enumerate(data[:,0]):
    cond = (t>=dt-15)&(t<dt+15)
    
    rpredict_sun[i] = np.mean(1/dtheory_sun[cond]**2)
    rpredict_jup[i] = np.mean(1/dtheory_jup[cond]**2)
```

```python
#Log-Likelihood is defined and the MCMC is running

def LogLikelihood(n_sun,n_jup,n_tot):
    prediction = n_sun*(rpredict_sun) + n_jup*(rpredict_jup) + n_tot
    return -0.5*(np.sum(np.log10(2*np.pi*data[:,2]**2) + (prediction-data[:,1])**2/data[:,2]**2))



info = {"likelihood": {"Chi2": LogLikelihood},
        
        "params": dict([
                        ("n_sun" , {"prior" : {"dist": "norm", "loc": 25, "scale" : 2},
                                    "latex" : r"R_{sun}"  }),
                        
                        ("n_jup" , {"prior" : {"min": 0, "max": 5},
                                    "latex": r"\dfrac{R_{jup}}{(5 \ \rm AU)^2} \ \rm (cpd/100t)"  }),
            
                       
                        ("n_tot" , {"prior" : {"min": 0, "max": 30},
                                    "latex": r"R_{B} \ \rm (cpd/100t)"  })                        
                        ]),
        
        "sampler": {"mcmc": {"Rminus1_stop": 0.001, "max_tries": 100000}}
       }


updated_info,sampler = run(info)
```

```python
# Filled 2D comparison plot with legend
gd_sample = MCSamplesFromCobaya(updated_info, sampler.products()["sample"])

g = plots.get_single_plotter(width_inch=5, ratio=1)
g.plot_2d(gd_sample, "n_jup","n_tot", filled=True,lims=[0, 4, 0, 13])
#g.export('./bayes.pdf')
```

```python
#best fit and covariance matrix and errors
mean   = gd_sample.getMeans()
covmat = gd_sample.getCovMat().matrix[:, :]
print("Mean:")
print(mean)
print("Covariance matrix:")
print(covmat)
print('####################')

stats = gd_sample.getMargeStats()
lims1 = stats.parWithName('n_sun').limits
lims2 = stats.parWithName('n_tot').limits
lims3 = stats.parWithName('n_tot').limits

for conf, lim1, lim2, lim3 in zip(gd_sample.contours, lims1,lims2,lims3):
    print('nsun %s%% lower: %.1f upper: %.1f (%s)'%(conf, lim1.lower, lim1.upper, lim1.limitType()))
    print('njup %s%% lower: %.2f upper: %.2f (%s)'%(conf, lim2.lower, lim2.upper, lim2.limitType()))
    print('ntot %s%% lower: %.2f upper: %.2f (%s)'%(conf, lim3.lower, lim3.upper, lim3.limitType()))
    print('####################')
```

```python
#test statistics Analysis

n               = 30000
prediction_sun1 = 24.6*(rpredict_sun)*30
background_sun  = 9.2*30

prediction_sun2 = 25.3*(rpredict_sun)*30
prediction_jup  = 1.5*(rpredict_jup)*30 
background_jup  = 7.0*30



sample_1 = np.zeros((n,len(data[:,2])))
sample_2 = np.zeros((n,len(data[:,2])))
for i in range(len(data[:,2])):
    np.random.seed(i)
    sample_1[:,i] = np.random.poisson(prediction_sun1[i],n) + np.random.poisson(background_sun,n)
    np.random.seed(i+1456)
    sample_2[:,i] = np.random.poisson(prediction_sun2[i],n) + np.random.poisson(prediction_jup[i],n) + np.random.poisson(background_jup,n)
    
chi_sample_1 = (np.sum((sample_1- prediction_sun1 - background_sun)**2/(prediction_sun1 + background_sun),axis=1)-
                np.sum((sample_1- prediction_sun2 - prediction_jup - background_jup)**2/(prediction_sun2 + prediction_jup + background_jup),axis=1))

chi_sample_2 = (np.sum((sample_2- prediction_sun1 - background_sun)**2/(prediction_sun1 + background_sun),axis=1)-
                np.sum((sample_2- prediction_sun2 - prediction_jup - background_jup)**2/(prediction_sun2 + prediction_jup + background_jup),axis=1))


t_real = (np.sum((30*data[:,1]- prediction_sun1 - background_sun)**2/(prediction_sun1 + background_sun))-
                np.sum((30*data[:,1]- prediction_sun2 - prediction_jup - background_jup)**2/(prediction_sun2 + prediction_jup + background_jup)))


print('t stat:')
print(t_real)
print()
```

```python
print(100*(1-len(chi_sample_1[chi_sample_1>=t_real])/n))
print(100*(1-len(chi_sample_2[chi_sample_2>=t_real])/n))

plt.hist(chi_sample_2,bins=70,alpha=0.7,color='b',label=r'$\rm true \ JUP+SUN $')
plt.hist(chi_sample_1,bins=70,alpha=0.5,color='r',label=r'$\rm true \ SUN $')

plt.vlines(x=t_real,ymin=0,ymax=1400,lw=3,ls='--',color='y',label=r'$\rm BOREXINO \ result$')
plt.xlabel(r'${t} = \chi^2_{\rm sun} - \chi^2_{\rm jup+sun}$')
plt.ylabel(r'$\rm Entries$')

plt.legend(loc='upper left',fontsize=14)
#plt.savefig('./t_stat.pdf',bbox_inches='tight')
```
